Jelaskan perbedaan antara Univariate, Bivariate dan Multivariate Analysis 

Jawaban:
Univariate = Imbuhan awal ‘Uni’ artinya ‘satu’, digunakan untuk analisis data feature tunggal merupakan sebuah teknik dalam memahami dan melakukan eksplorasi data. 

Bivariate = Imbuhan awal 'Bi' artinya 'dua', digunakan untuk untuk menganalisis 2 variables dan menemukan sebuah relasi. Analisis bivariate juga merupakan salah satu cara untuk menggunakan koefisien korelasi dalam rangka menemukan apakah dua variabel memiliki relasi atau tidak.

Multivariate = untuk menganalisis lebih dari 2 variabel di waktu yang sama, trends yang dihasilkan bisa menjadi multidimensi secara alami, dengan analisis ini akan membantu kita memahami manakah data yang memiliki tren yang kompleks pada kombinasi atribut.

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g

```python
#code here
# Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm
GroupSpecies.plot(kind='pie', 
                  y = 'flipper_length_mm',
                  autopct='%1.1f%%', #%1.0f%% = Persen dibulatkan keatas
                  legend=False)
plt.show()
```

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here
# Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar'
CatPlot = sns.catplot(x='sex', y='bill_length_mm', kind='bar', data=df)
CatPlot
```

Analisis =

Dari hasil uji dengan catplot didapat jenis kelamin Male memiliki nilai bill_length_mm lebih besar dari jenis kelamin Female.

```python
# Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm'
ScatterPlot = sns.scatterplot(x='body_mass_g',
                    y='flipper_length_mm',
                    hue='bill_length_mm',
                    data=df)
ScatterPlot
```

Analisis =

Dari hasil uji dengan scatterpot, data dari bill_length_mm bersifat positif. semakin tinggi nilai body_mass_g semakin tinggi juga nilai flipper_length_mm.
