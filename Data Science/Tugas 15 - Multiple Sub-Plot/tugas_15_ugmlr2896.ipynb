<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
# Grafik CosX
fig, ax = plt.subplots(figsize=(10,8)) #(Lebar, Tinggi)
ax.plot(x, y)
ax.set_title('Grafik CosX')

# SubGrafik CosX**2
ins1 = ax.inset_axes([-0.2, -0.1, 0.3, 0.3])#[Geser Kanan,Geser Atas, Tarik Kanan, Tarik Atas]
ins1.plot(x, y)
ins1.plot(x, y2, c='r')
ins1.set_title('Grafik CosX^2')

# SubGrafik CosX**3
ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins2.plot(x, y)
ins2.plot(x, y3, c='g')
ins2.set_title('Grafik CosX^3')

# SubGrafik CosX**4
ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins3.plot(x, y)
ins3.plot(x, y4, c='black')
ins3.set_title('Grafik CosX^4')

# SubGrafik CosX**5
ins4 = ax.inset_axes([-0.2, 0.9, 0.3, 0.3])
ins4.plot(x, y)
ins4.plot(x, y5, c='y')
ins4.set_title('Grafik CosX^5')

plt.show()

```
