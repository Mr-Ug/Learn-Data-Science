```python
# code here
# 1. Mengganti fields yang berisi nilai error ('Na','-','blank','...',dsb) ke nilai kosong (NaN)
missing_values = ['na','-']
Data_Clean = pd.read_csv('austin_weather.csv', na_values=missing_values)


# 2. Mengubah tipe data (Object) menjadi Float
Data_Clean.astype({'DewPointAvgF': 'float', 'HumidityAvgPercent': 'float', 'WindAvgMPH': 'float'}).dtypes
# print(Data_Clean.isna().sum()) # .isna().sum() menghitung banyaknya fields yang kosong (NaN)


# 3. Mengisi fields kosong (NaN) dengan nilai mean
peng_no_NaN = Data_Clean.fillna(value=Data_Clean.mean())
# print(peng_no_NaN.isna().sum())
```

Analisa :

```python
DF_Sample1000 = Data_Clean.sample(n=1000)
# DF_Sample1000
plt.scatter(DF_Sample1000['TempAvgF'], DF_Sample1000['WindAvgMPH'], s=DF_Sample1000['TempAvgF'], cmap='coolwarm')
plt.show()
```

Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut
*   Remove data outliers tersebut menggunakan batas atas dan batas bawah

```python
print("The shape of the dataframe is: ", df2.shape)
df2.describe()
```

```python
sns.boxplot(x=df2['House_Price'])
plt.show()
```

```python
# df2["House_Price"] = df2["House_Price"].astype(float)

# Hitung nilai Q1 dan Q3
Q1 = np.nanpercentile(df2, 25)
Q3 = np.nanpercentile(df2, 75)

# Hitung nilai IQR
IQR = Q3 - Q1

print(f'Batas Atas = {Q1}')
print(f'Batas Bawah = {Q3}')
print(f'IQR = , IQR')
```
