<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

# Menampilkan Figure dan Axis Kosong
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# explicit
x = np.linspace(0, 20, 100)
y = np.cos(x/2)

fig, ax=plt.subplots()
plt.title("Cara Explicit")
ax.plot(x, y)

plt.show()
```

```python
# implicit
plt.title("Cara Implicit")
plt.plot(x, y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)

# - Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
plt.plot(x, y, color="red")
plt.plot(x, y2, color="green")
plt.plot(x, y3, color="blue")

plt.style.use('seaborn')
plt.savefig('GrafikSoal3.png')
plt.show()
```

```python
# - save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.
# Menampilkan gambar GrafikSoal3
# plt.savefig('GrafikSoal3.png')

img = cv2.imread('grafiksoal3.png')
plt.imshow(img)
plt.show()
```
